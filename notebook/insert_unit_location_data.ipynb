### 2. 데이터 전처리

```python
import pandas as pd
```

```python
df = pd.read_json('./locationinfoUnit.json')
df.head()
```

```python
df = df.transpose()
df.head()
```

```python
colums = [
    'route_id',
    'unit_name',
    'route_name',
    'latitude',
    'label',
    'longitude',
]
df.columns = colums
df = df.reset_index()
df = df.drop(columns=['index'])

df_cleaned = df.applymap(lambda x: x[0]["value"] if isinstance(x, list) and len(x) > 0 else None)
df_cleaned['longitude'] = df_cleaned['longitude'].astype(float)
df_cleaned['latitude'] = df_cleaned['latitude'].astype(float)
df_cleaned.head()
```

```python
print(df_cleaned.isnull().sum())
```

```python
# latitude가 0인 row 찾기
df_cleaned[df_cleaned['latitude'] == 0]
```

### 데이터 입력

```python
from dotenv import dotenv_values

config = dotenv_values(".env")
POSTGRES_USER = config.get('POSTGRES_USER')
POSTGRES_PASSWORD = config.get('POSTGRES_PASSWORD')
POSTGRES_HOST = config.get('POSTGRES_HOST')
POSTGRES_PORT = config.get('POSTGRES_PORT')
POSTGRES_DB = config.get('POSTGRES_DB')
```

```python
from sqlalchemy import create_engine

engine = create_engine(f'postgresql+psycopg2://{POSTGRES_USER}:{POSTGRES_PASSWORD}@{POSTGRES_HOST}:{POSTGRES_PORT}/{POSTGRES_DB}')
df_cleaned.to_sql(name='unit_location', con=engine, if_exists='replace', index=False)
```
