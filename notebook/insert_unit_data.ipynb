## 영업소 데이터 삽입

> URL: https://data.ex.co.kr/portal/fdwn/view?type=ETC&num=75&requestfrom=dataset

```python
import pandas as pd
import numpy as np
```

### 데이터 가져오기

```python
df = pd.read_csv('./영업소.csv')
df.head()
```

### 데이터 전처리

```python
columns = [
    "unit_id",
    "unit_name",
    "agency_code",
    "route_id",
    "center_code",
    "tcs_branch_code",
    "unit_type_code",
    "unit_method_code",
    "phone_number",
    "is_operational_office",
    "entrance_lane_count",
    "exit_lane_count",
    "variable_lane_count",
    "is_last_private_ic",
    "is_virtual_office"
]

df.columns = columns
df.head()
```

```python
df = df.applymap(lambda x: None if isinstance(x, str) and x.strip() == "" else x)
df = df.applymap(lambda x: None if isinstance(x, str) and x == np.nan else x)
df.head()
```

### 데이터 삽입

```python
from dotenv import dotenv_values

config = dotenv_values(".env")
POSTGRES_USER = config.get('POSTGRES_USER')
POSTGRES_PASSWORD = config.get('POSTGRES_PASSWORD')
POSTGRES_HOST = config.get('POSTGRES_HOST')
POSTGRES_PORT = config.get('POSTGRES_PORT')
POSTGRES_DB = config.get('POSTGRES_DB')
```

```python
from sqlalchemy import create_engine

engine = create_engine(f'postgresql://{POSTGRES_USER}:{POSTGRES_PASSWORD}@{POSTGRES_HOST}:{POSTGRES_PORT}/{POSTGRES_DB}')
df.to_sql(name='unit', con=engine, if_exists='append', index=False)
```
